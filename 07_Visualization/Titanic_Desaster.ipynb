```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("white")
```

```python
dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(dataset_url, sep=',')
titanic.head()
```

```python
titanic = titanic.set_index("PassengerId")
titanic.head()
```

```python
male = (titanic.Sex == 'male').sum()
female = (titanic.Sex == 'female').sum()
proportions = [male, female]

plt.pie(proportions, labels=['Male', 'Female'], colors=['blue', 'red'], shadow=False, explode=(0.15, 0), startangle=90)
plt.title("Sex Proportion")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Sex', data=titanic)
```

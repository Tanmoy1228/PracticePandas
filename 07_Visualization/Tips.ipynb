```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style("white")
```

```python
dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips = pd.read_csv(dataset_url, sep=',')
tips.head()
```

```python
del tips['Unnamed: 0']
tips.head()
```

```python
total_bill_hist = sns.distplot(tips.total_bill)
total_bill_hist.set(xlabel = 'Value', ylabel = 'Frequency', title = "Total Bill")
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.scatterplot(x="tip", y="day", hue="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
```

```python
sns.set(style="ticks")
g = sns.FacetGrid(data=tips, col="time")
g.map(plt.hist, "tip")
```

```python
g = sns.FacetGrid(data=tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip")
```

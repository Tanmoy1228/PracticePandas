```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

%matplotlib inline
```

```python
dataset_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(dataset_url, sep='\t')
chipo.head()
```

```python
chipo.head(10)
```

```python
# item_name = chipo.item_name
# letter_counts = Counter(item_name)
# df = pd.DataFrame.from_dict(letter_counts, orient='index')
# df = df[0].sort_values(ascending=False)[0:5]

df = chipo.item_name.value_counts()[0:5]

df.plot(kind='bar')
plt.xlabel('Items')
plt.ylabel('Number of Times ordered')
plt.title('Most ordered items.')
plt.show()
```

```python
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price]
df = chipo.groupby(by='order_id').sum()
plt.scatter(x = df.item_price, y = df.quantity, s=50, c='green')

plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
```

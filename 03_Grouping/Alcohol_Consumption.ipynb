```python
import pandas as pd
```

```python
dataset_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(dataset_url, sep=',')
drinks.columns
```

```python
drinks.groupby(by='continent').mean().sort_values(by='beer_servings', ascending=False).beer_servings
```

```python
continent = drinks.groupby('continent')
continent.wine_servings.describe()
```

```python
drinks.groupby('continent').mean()
```

```python
drinks.groupby('continent').median()
```

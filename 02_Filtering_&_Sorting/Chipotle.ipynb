```python
import pandas as pd
```

```python
dataset_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(dataset_url, sep='\t')
```

```python
price = [float(value[1: -1]) for value in chipo.item_price]
chipo.item_price = price
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])
chipo_filtered = chipo_filtered[chipo_filtered.quantity == 1]
chipo_filtered.head(5)
```

```python
chipo_filtered = chipo_filtered.drop_duplicates(['item_name'])
chipo_filtered = chipo_filtered[['item_name', 'item_price']]
chipo_filtered = chipo_filtered.sort_values(by='item_price', ascending=False)
chipo_filtered
```

```python
chipo_shorted_by_name = chipo.sort_values(by='item_name', ascending=True)
chipo_shorted_by_name
```

```python
chipo_shorted_by_price = chipo.sort_values(by='item_price', ascending=False)
chipo_shorted_by_price.head(1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
apple = pd.read_csv(dataset_url, sep=',')
apple.head()
```

```python
apple.dtypes
```

```python
apple.Date = pd.to_datetime(apple.Date)
apple.dtypes
```

```python
apple = apple.set_index('Date')
apple.head()
```

```python
apple.index.is_unique
```

```python
apple = apple.sort_index(ascending=True)
apple.head()
```

```python
apple_month = apple.resample('BM').mean()
apple_month.head()
```

```python
apple.index.max() - apple.index.min()
```

```python
len(apple.resample('M').mean())
```

```python
apple_plt = apple['Adj Close'].plot(title = "Apple Stock")

fig = apple_plt.get_figure()
fig.set_size_inches((13.5, 9))
fig.show()
```

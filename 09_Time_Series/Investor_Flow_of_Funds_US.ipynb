```python
import pandas as pd
import datetime as dt
```

```python
dataset_url = 'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv'
df = pd.read_csv(dataset_url, sep=',')
df.head()
```

```python
df = df.set_index('Date')
df.head()
```

```python
df.index
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
monthly = df.resample('M').sum()
monthly.head()
```

```python
monthly = monthly.dropna()
monthly.head()
```

```python
yearly = monthly.resample('AS').sum()
yearly.head()
```

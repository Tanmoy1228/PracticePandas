```python
import pandas as pd
```

```python
dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(dataset_url, sep=',')
df.columns
```

```python
df = df.loc[:, 'school': 'guardian']
df.head()
```

```python
capitalizer = lambda x: x.capitalize()
```

```python
df['Mjob'].apply(capitalizer)
df['Fjob'].apply(capitalizer)
```

```python
df.tail(1)
```

```python
df['Mjob'] = df['Mjob'].apply(capitalizer)
df['Fjob'] = df['Fjob'].apply(capitalizer)
df.tail()
```

```python
def is_legal_drinker(age):
    if age > 17:
        return True
    return False

df['legal_drinker'] = df['age'].apply(is_legal_drinker)
df.head()
```

```python
def mul10(x):
    if type(x) is int:
        x = x*10
    return x
df = df.applymap(mul10)
df.head()
```

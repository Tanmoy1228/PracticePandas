```python
import pandas as pd
```

```python
dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'
crime = pd.read_csv(dataset_url, sep=',')
crime.head()
```

```python
crime.info()
```

```python
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

```python
crime = crime.set_index(crime.Year, drop=True)
crime.head()
```

```python
del crime['Total']
crime.head()
```

```python
crime_decade = crime.resample('10AS').sum()
population = crime['Population'].resample('10AS').max()
crime_decade.Population = population
crime_decade
```

```python
import pandas as pd
```

```python
dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'
baby_names = pd.read_csv(dataset_url, sep=',')
baby_names.columns
```

```python
baby_names.head(10)
```

```python
baby_names = baby_names[['Name', 'Year', 'Gender', 'State', 'Count']]
baby_names.head()
```

```python
baby_names.Gender.value_counts()
```

```python
names = baby_names.groupby("Name").sum()
del names['Year']
names.sort_values(by='Count', ascending=False).head()
```

```python
len(names)
```

```python
names.Count.idxmax()
```

```python
len(names[names.Count == names.Count.min()])
```

```python
names[names.Count == names.Count.median()]
```

```python
names.Count.std()
```

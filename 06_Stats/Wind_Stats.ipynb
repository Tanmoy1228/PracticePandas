```python
import pandas as pd
import datetime
```

```python
dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(dataset_url, sep='\s+', parse_dates=[[0, 1, 2]])
data.head()
```

```python
def fix_year(date):
    year = date.year
    if date.year > 1989:
        year = date.year - 100
    return datetime.date(year, date.month, date.day)
data.Yr_Mo_Dy = data.Yr_Mo_Dy.apply(fix_year)
data.head()
```

```python
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
data = data.set_index('Yr_Mo_Dy')
data.head()
```

```python
day_stats = pd.DataFrame()

day_stats['min'] = data.min(axis = 1) 
day_stats['max'] = data.max(axis = 1) 
day_stats['mean'] = data.mean(axis = 1) 
day_stats['std'] = data.std(axis = 1) 

day_stats.head()
```

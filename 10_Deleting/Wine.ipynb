```python
import pandas as pd
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine = pd.read_csv(url)

wine.head()
```

```python
wine = wine.drop(wine.columns[[0, 3, 6, 8, 10, 12, 13]], axis=1)
wine.head()
```

```python
wine.columns = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
wine.head()
```

```python
wine.iloc[0:3, 0:1] = np.nan
wine.head()
```

```python
wine.iloc[2:4, 3:4] = np.nan
wine.head()
```

```python
wine.alcohol.fillna(10, inplace=True)
wine.magnesium.fillna(100, inplace=True)
wine.head()
```

```python
wine.isnull().sum()
```

```python
random = np.random.randint(0, 10, size=10)
random.sum()
```

```python
wine.alcohol[random] = np.nan
wine.head(10)
```

```python
wine.isnull().sum()
```

```python
wine = wine.dropna(how='any')
wine.head()
```

```python
wine.alcohol.notnull()
```

```python
wine = wine.reset_index()
wine.head()
```
